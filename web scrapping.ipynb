# # import library

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

# setting scrapping website

```python
url = "https://www.flipkart.com/search?q=ear%20pods&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
allow = requests.get(url)
```

```python
print(allow)
```

# Testing scrapping allowability

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

### Scrapping

```python
product = soup.select(".s1Q9rs")
```

```python
product
```

```python
product_name = []
```

```python
for item in product:
    product_name.append(item.text)
```

```python
product_name
```

```python
prices = soup.select("._8VNy32 ._30jeq3")
```

```python
prices
```

```python
cost = []
```

```python
for item in prices:
    cost.append(item.text)
```

```python
cost
```

### creating dataframe

```python
import pandas as pd
```

```python
table = pd.DataFrame(product_name,columns=['product'])
```

```python
table
```

```python
table['cost'] = cost
```

```python
table
```

```python
table.to_csv('EarPods.csv')
```
